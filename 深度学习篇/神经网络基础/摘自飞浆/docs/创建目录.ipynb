```python
import os

def generate_markdown_directory(path, indent=0):
    """
    遍历指定路径及其子目录，输出Markdown格式的目录结构。
    :param path: 要遍历的目录路径。
    :param indent: 缩进级别，用于格式化输出。
    """
    try:
        files_and_dirs = sorted(os.listdir(path))  # 获取所有文件和目录，进行排序
        for item in files_and_dirs:
            full_path = os.path.join(path, item)
            if os.path.isdir(full_path):
                # 如果是目录，则递归调用，并增加缩进级别
                print(f"{'    '*indent}- **{item}**")
                generate_markdown_directory(full_path, indent + 1)
            elif item.endswith('.md'):
                # 如果是.md文件，则直接输出
                print(f"{'    '*indent}- [{item}]({path}/{item})")
                print("\n")
    except PermissionError:
        # 如果没有权限访问某个目录，则跳过
        pass

# 调用函数，从当前目录开始遍历
generate_markdown_directory('/home/charles/code/about_Radar/深度学习篇/神经网络基础/摘自飞浆/docs')
```
