```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# morlet小波基函数

t = np.linspace(-8, 8, 1000)

w0 = 2
real_ = np.cos(2*np.pi*w0*t)*np.exp(-0.5*(np.power(t, 2)))
img_ = np.sin(2*np.pi*w0*t)*np.exp(-0.5*(np.power(t, 2)))

freq_ = np.abs(np.fft.fft(real_))

plt.figure(figsize=(8, 2))
plt.subplot(131)
plt.plot(real_)
plt.subplot(132)
plt.plot(img_)
plt.subplot(133)
plt.plot(freq_[0:100])

w0 = 5
real_ = np.cos(2*np.pi*w0*t)*np.exp(-0.5*(np.power(t, 2)))
img_ = np.sin(2*np.pi*w0*t)*np.exp(-0.5*(np.power(t, 2)))

freq_ = np.abs(np.fft.fft(real_))

plt.figure(figsize=(8, 2))
plt.subplot(131)
plt.plot(real_)
plt.subplot(132)
plt.plot(img_)
plt.subplot(133)
plt.plot(freq_[0:100])
```
