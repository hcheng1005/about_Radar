```python
import numpy as np
import matplotlib.pyplot as plt

# 模拟相位数据
x = np.linspace(0, 10, 100)
phase_original = np.sin(x * 2 * np.pi / 3) * 2 * np.pi  # 模数在 [-2π, 2π]

# 添加周期性相位跳变
phase_wrapped = np.mod(phase_original + np.pi, 2 * np.pi) - np.pi

# 相位解缠
def phase_unwrap(phase):
    unwrapped = np.unwrap(phase)
    return unwrapped

# 解缠后的相位
phase_unwrapped = phase_unwrap(phase_wrapped)

# 可视化结果
plt.figure(figsize=(12, 6))
plt.subplot(3, 1, 1)
plt.plot(x, phase_original, label="Original Phase")
plt.title("Original Phase")
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(x, phase_wrapped, label="Wrapped Phase")
plt.title("Wrapped Phase")
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(x, phase_unwrapped, label="Unwrapped Phase")
plt.title("Unwrapped Phase")
plt.legend()

plt.tight_layout()
plt.show()
```
